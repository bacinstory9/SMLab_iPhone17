```python
import pandas as pd
import os
from color_mapping import clean_color_column
from country_mapping import clean_country_column

file_path = os.path.expanduser("~/Desktop/regge.xlsx")
df = pd.read_excel(file_path)

# Clean the color column
df["Farbe"] = clean_color_column(df["Farbe"])

# Clean the condition column
df["Zustand"] = clean_color_column(df["Zustand"])
```

```python
categorical_columns = [
    "Zustand",
    "Farbe",
    "Seller Type",
    "Herkunftsland",
    "Verkaufsart",
    "Batteriezustand"
]

tables = {}
def frequency_table(data, column):
    # dropna=False zählt auch die fehlenden Werte
    count = data[column].value_counts(dropna=False)
    percent = data[column].value_counts(dropna=False, normalize=True) * 100

    return pd.DataFrame({
        'Anzahl': count,
        'Prozent': percent.round(2)
    })


# Jetzt funktioniert dein ursprünglicher Code:
for col in categorical_columns:
    tables[col] = frequency_table(df, col)
    print("\n", "="*50)
    print(col)
    display(tables[col])
```
